```python
import numpy as np

import scienceplots
import matplotlib.pyplot as plt
plt.style.use('science')


from CML_tool.xAI_plots.model_explanations import plot_shap_ridgelines, plot_comparing_shap_ridgelines, plot_comparing_raw_data_ridgelines
from CML_tool.xAI_plots.feature_maps import SLE_2021
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate random means and standard deviations for 50 columns
means_1 = np.random.uniform(0, 1, 50)
stds_1 = np.random.uniform(0, 0.1, 50)
means_2 = np.random.uniform(0, 1, 50)
stds_2 = np.random.uniform(0, 0.1, 50)

# Generate two (500, 50) matrices
matrix_1 = np.array([np.random.normal(mean, std, 500) for mean, std in zip(means_1, stds_1)]).T
matrix_2 = np.array([np.random.normal(mean, std, 500) for mean, std in zip(means_2, stds_2)]).T
```

```python
FEATURE_MAP = {f'S-{x}':f'f{x}' for x in range(50)}
```

```python
fig1 = plot_shap_ridgelines(
    shaps=matrix_1,
    all_features_names=[f'S-{x}' for x in range(50)],
    data_rep='Signatures',
    features_name_map=FEATURE_MAP,
    title='Test',
    xmax=1.5,
    bandwidth=0.002,
    figsize=(5,3),
    overlap=0.75
)
```

```python
fig2 = plot_comparing_shap_ridgelines(
    shaps_ref=matrix_1,
    shaps_comp=matrix_2,
    label_ref='reference',
    label_comp='comparison',
    all_features_names=[f'S-{x}' for x in range(50)],
    data_rep='Signatures',
    features_name_map=FEATURE_MAP,
    title='Test',
    xmax=1.75,
    bandwidth=0.02,
    figsize=(5,3),
    overlap=0.6,
    color_kde_ref='royalblue',
    color_rugs_ref='slateblue',
    color_kde_comp='orangered',
    color_rugs_comp='lightsalmon',
)
```

```python
fig3 = plot_comparing_shap_ridgelines(
    shaps_ref=matrix_1,
    shaps_comp=matrix_2,
    label_ref='reference',
    label_comp='comparison',
    all_features_names=[f'S-{x}' for x in range(50)],
    data_rep='Signatures',
    features_name_map=FEATURE_MAP,
    title='Test',
    xmax=1.75,
    bandwidth=0.02,
    figsize=(5,3),
    overlap=0.6,
    color_kde_ref='royalblue',
    color_rugs_ref='slateblue',
    color_kde_comp='orangered',
    color_rugs_comp='lightsalmon',
    order = [49,1,22,33,4,5,6,7,8,9,10],
    top=10
)
```
